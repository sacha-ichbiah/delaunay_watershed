```python
val_thresh=0.72
point_density = 0.6
```

```python
DT = cell_prob-np.amin(cell_prob)
DT/=np.amax(DT)
DT=1-DT
Total_EDT,dist_points,threshold_on_EDT = auto_double_edt(DT,val_thresh,normalize=False)
dist_points/=point_density
min_dist = dist_points.astype(int)
distance_thresh = min_dist

mask_1 = DT>val_thresh
mask_1 = np.pad(mask_1, ((1, 1), (1, 1)), 'constant',constant_values = 0)
points,idx_max_interior_points = double_filtering(Total_EDT,mask_1,threshold_on_EDT,distance_thresh=distance_thresh,min_dist=min_dist)
idx_interior_points = np.arange(idx_max_interior_points)

points_f=points.astype(float)
points_f[:,0]/=Total_EDT.shape[0]
points_f[:,1]/=Total_EDT.shape[0]
tesselation=Delaunay(points_f)

alpha = Total_EDT.shape[1]/Total_EDT.shape[0]
fig,axs=plt.subplots(1,3,figsize=(15,5))
points
axs[0].imshow(DT>val_thresh)
#axs[1].imshow(Total_EDT,extent=(0.,alpha,0.,1))
axs[1].triplot(points[:,1], 1-points[:,0], tesselation.simplices)

import skimage.filters
Total_EDT = skimage.filters.gaussian(Total_EDT,sigma=1)
Total_EDT/=np.amax(Total_EDT)
Seg = Plotting_Wrapper(tesselation,Total_EDT)

Seg.Delaunay_Graph.cut_edges(threshold_on_EDT)
#fig=Seg.plot_triangulation(image=True)
#plt.figure()
Seg.build_graph(highest_removed=False)
Centroids = Seg.Delaunay_Graph.compute_nodes_centroids()
EDT = Seg.image
Coords = np.array([[15,15],[75,125],[150,175],[150,125],[220,175]])
idx_seeds = compute_seeds_idx_from_pixel_coords(EDT,Centroids,Coords)
Seg.build_graph(highest_removed=False)
Seg.watershed_seeded(idx_seeds)
Seg.combine_maps()
fig=Seg.plot_polylines(figsize=10,polylines=True)

from Dcel import * 
DCEL = DCEL_Data(Seg)
Angles_dict = DCEL.compute_angles()
Curvature_dict = DCEL.compute_curvatures()
Length_dict = DCEL.compute_lengths()



import pandas as pd
excel_file = "Segmentation_quadruplet.xlsx"
Excel_file = pd.read_excel(excel_file,sheet_name='mov_1')
Excel_file['Time'][i]=i
Excel_file['Length ABA-0'][i]=Length_dict[(0,1)]
Excel_file['Length ABP-0'][i]=Length_dict[(0,2)]
Excel_file['Length EMS-0'][i]=Length_dict[(0,3)]
Excel_file['Length P2-0'][i]=Length_dict[(0,4)]
Excel_file['Length ABA-ABP'][i]=Length_dict[(1,2)]
Excel_file['Length ABA-EMS'][i]=Length_dict[(1,3)]
Excel_file['Length ABP-EMS'][i]=Length_dict[(2,3)]
Excel_file['Length ABP-P2'][i]=Length_dict[(2,4)]   ###
Excel_file['Length EMS-P2'][i]=Length_dict[(3,4)]
Excel_file['Curvature ABA-0'][i]=Curvature_dict[(0,1)]
Excel_file['Curvature ABP-0'][i]=Curvature_dict[(0,2)]
Excel_file['Curvature EMS-0'][i]=Curvature_dict[(0,3)]
Excel_file['Curvature P2-0'][i]=Curvature_dict[(0,4)]
Excel_file['Curvature ABA-ABP'][i]=Curvature_dict[(1,2)]
Excel_file['Curvature ABA-EMS'][i]=Curvature_dict[(1,3)]
Excel_file['Curvature ABP-EMS'][i]=Curvature_dict[(2,3)]
Excel_file['Curvature ABP-P2'][i]=Curvature_dict[(2,4)] 
Excel_file['Curvature EMS-P2'][i]=Curvature_dict[(3,4)]
Excel_file['Angles 0-ABA-ABP'][i]=Angles_dict[(0, 1, 2)]*180/np.pi
Excel_file['Angles 0-ABP-P2'][i]=Angles_dict[(0, 2, 4)]*180/np.pi
Excel_file['Angles 0-EMS-ABA'][i]=Angles_dict[(0, 3, 1)]*180/np.pi
Excel_file['Angles 0-P2-EMS'][i]=Angles_dict[(0, 4, 3)]*180/np.pi
Excel_file['Angles ABA-0-EMS'][i]=Angles_dict[(1, 0, 3)]*180/np.pi
Excel_file['Angles ABA-ABP-0'][i]=Angles_dict[(1, 2, 0)]*180/np.pi
Excel_file['Angles ABA-EMS-ABP'][i]=Angles_dict[(1, 3, 2)]*180/np.pi
Excel_file['Angles ABP-0-ABA'][i]=Angles_dict[(2, 0, 1)]*180/np.pi
Excel_file['Angles ABP-ABA-EMS'][i]=Angles_dict[(2, 1, 3)]*180/np.pi
Excel_file['Angles ABP-EMS-P2'][i]=Angles_dict[(2, 3, 4)]*180/np.pi
Excel_file['Angles ABP-P2-0'][i]=Angles_dict[(2, 4, 0)]*180/np.pi
Excel_file['Angles EMS-0-P2'][i]=Angles_dict[(3, 0, 4)]*180/np.pi
Excel_file['Angles EMS-ABA-0'][i]=Angles_dict[(3, 1, 0)]*180/np.pi
Excel_file['Angles EMS-ABP-ABA'][i]=Angles_dict[(3, 2, 1)]*180/np.pi
Excel_file['Angles EMS-P2-ABP'][i]=Angles_dict[(3, 4, 2)]*180/np.pi
Excel_file['Angles P2-0-ABP'][i]=Angles_dict[(4, 0, 2)]*180/np.pi
Excel_file['Angles P2-ABP-EMS'][i]=Angles_dict[(4, 2, 3)]*180/np.pi
Excel_file['Angles P2-EMS-0'][i]=Angles_dict[(4, 3, 0)]*180/np.pi

Excel_file['Val_thresh'][i]=val_thresh
Excel_file['Point_density'][i]=point_density

writer = pd.ExcelWriter(excel_file)#mode = 'a',engine="openpyxl")# engine='xlsxwriter',mode='a')
Excel_file.to_excel(writer, sheet_name='mov_1', index=False)
writer.save()
Excel_file[i-2:i+1]


```
