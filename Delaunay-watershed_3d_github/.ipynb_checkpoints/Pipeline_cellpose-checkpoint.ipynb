```python
import skimage.io as io
from cellpose import models, io
import os
import numpy as np
from skimage import img_as_ubyte


d = 100
folder_mov = "Celegans_data/cellmembrane_GFP/inside_eggshell/"
names =["190429-2",
        "190429-3",
        "190429-4",
        "190820-2",
        "190820-3"]

for name in names : 
    try : os.mkdir("Results_C_elegans/inside_shell/Results_"+name)
    except : pass
    
    Mov = io.imread(folder_mov+name+".tif")
    print(Mov.shape)
    
    for t in range(Mov.shape[0]):

        Stack = Mov[t,0,:,:].copy().astype(float)
        Stack-=np.amin(Stack)
        Stack/=np.amax(Stack)

        Stack = img_as_ubyte(Stack)
        
        model = models.Cellpose(gpu=True, model_type='cyto')
        masks, flows, styles, diams = model.eval(Stack, diameter=d, channels=[0,0])

        io.masks_flows_to_seg(Stack, masks, flows, diams, "Results_C_elegans/inside_shell/Results_"+name+"/time_"+str(t), [0,0])
```

```python
import skimage.io as io
from cellpose import models, io
import os
import numpy as np
from skimage import img_as_ubyte


d = 100
folder_mut = "Celegans_data/cellshape_mutants/"
for folder in os.listdir(folder_mut): 

    folder_name = folder_mut+folder + "/"
    try : os.mkdir("Results_C_elegans/mutants/"+folder)
    except : print("non")

    for name in os.listdir(folder_name): 
        mov_name = name[:-4]
        try : os.mkdir("Results_C_elegans/mutants/"+folder + "/"+ mov_name)
        except : print("non plus")
        Mov = io.imread(folder_name+mov_name+".tif")
        print(Mov.shape)
        for t in range(Mov.shape[0]):

            Stack = Mov[t,:,:].copy().astype(float)
            Stack-=np.amin(Stack)
            Stack/=np.amax(Stack)

            Stack = img_as_ubyte(Stack)

            model = models.Cellpose(gpu=True, model_type='cyto')
            masks, flows, styles, diams = model.eval(Stack, diameter=d, channels=[0,0])

            io.masks_flows_to_seg(Stack, masks, flows, diams, "Results_C_elegans/mutants/"+folder+"/"+mov_name+"/time_"+str(t), [0,0])

```
